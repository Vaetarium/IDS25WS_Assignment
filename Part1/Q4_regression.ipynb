# Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import seaborn as sns
```

```python
train_dataset_oh = pd.read_csv('./train_insurance_one_hot.csv')
x_train = train_dataset_oh.drop(columns=['id','charges','chargeGroup'], inplace=False)
y_train = train_dataset_oh[["charges"]]

test_dataset_oh = pd.read_csv('./test_insurance_one_hot.csv')
x_test = test_dataset_oh.drop(columns=['id','charges','chargeGroup'], inplace=False)
y_test = test_dataset_oh[["charges"]]
```

## a)

```python
# Plot charges over the age with implot function
sns.lmplot(x='age', y='charges', data=train_dataset_oh, height=6, aspect=1.5, line_kws={'color':'red'})
plt.title('Charges over the Age')
plt.show()
```

## b)

```python
# Linear Regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Print coefficients
coefficients = pd.DataFrame({
    'Feature': x_train.columns,
    'Coefficient': model.coef_[0]
})
print("Model Coefficients:\n", coefficients)

# MAE on training set
y_train_pred = model.predict(x_train)
mae_train = mean_absolute_error(y_train, y_train_pred)
print("\nMean Absolute Error on Training Set:", round(mae_train,2))

# MAE on test set
y_test_pred = model.predict(x_test)
mae_test = mean_absolute_error(y_test, y_test_pred)
print("Mean Absolute Error on Test Set:", round(mae_test,2))
```
